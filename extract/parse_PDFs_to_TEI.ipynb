```python
from grobid_client.grobid_client import GrobidClient


def parse_sample_pdfs(process_dir, tei_dir):
    client = GrobidClient(config_path="./config.json")
    client.process("processFulltextDocument", 
                   process_dir, 
                   output=tei_dir, 
                   consolidate_citations=1,
                   include_raw_citations=1,
                   segment_sentences=1, 
                   n=1)
```

```python
import glob
import os
from collections import Counter


def remove_process_pdfs(process_dir):
    process_files = glob.glob(os.path.join(process_dir, '*.pdf'))
    for process_file in process_files:
        !rm $process_file


def copy_pdfs(pdf_dir, process_dir, pdf_files):
    for pdf_file in pdf_files:
        process_file = pdf_file.replace(pdf_dir, process_dir)
        if process_file == pdf_file:
            raise ValueError(f"process_file has same name as pdf_file: '{process_file}' and '{pdf_file}'")
        !cp $pdf_file $process_file
        
        


pdf_files = glob.glob(os.path.join(pdf_dir, '*.pdf'))

step_size = 10
start = 10
for i in range(start, len(pdf_files), step_size):
    sample_files = pdf_files[i:i+step_size]
    print(i, i+step_size, len(sample_files))

    remove_process_pdfs(process_dir)
    copy_pdfs(pdf_dir, process_dir, sample_files)
    parse_sample_pdfs(process_dir, tei_dir)
```
